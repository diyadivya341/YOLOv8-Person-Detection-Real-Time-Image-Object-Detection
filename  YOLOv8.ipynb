```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load YOLOv8 model
model = YOLO("yolov8n.pt")  # or yolov8s.pt

# Read the local image (make sure sample.jpg is in your folder)
image_path = "mummy.jpg"
image = cv2.imread(r"C:\Users\divya\OneDrive\Pictures\mummy.jpg")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Perform object detection
results = model(image_rgb)   

# Draw bounding boxes
results[0].plot()

# Show output in Jupyter
plt.imshow(image_rgb)
plt.axis('off')
plt.title("YOLOv8 Detection Result")
plt.show()
```

```python
import matplotlib.pyplot as plt

results[0].plot()
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("YOLOv8 Object detection")
plt.show()
```

```python
for result in results:
    for box in result.boxes:
        confidence = box.conf.item()
        if confidence > 0.5: 
            class_id = int(box.cls)
            label = result.names[class_id]
            print(f"Detected: {label} ({confidence:.2f})")
```
